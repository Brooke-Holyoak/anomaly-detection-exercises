```python
def get_lower_and_upper_bounds(col, multiplier):
    

    #start with an inner fence calculation
    multiplier = 1.5
    #calculate q1 and q3
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    #calculate the iqr
    iqr = q3 - q1
    #calculate the inner lower and inner upper fences
    inner_lower_fence = q1 - (multiplier * iqr)
    inner_upper_fence = q3 + (multiplier * iqr)
    
    return inner_lower_fence, inner_upper_fence
```

```python
col = ('Temperature')

inner_lower_fence, inner_upper_fence = get_lower_and_upper_bounds(col, 1.5)
```
